```python
import psycopg2 as ps
import pandas as pd
import numpy as np
```

```python
def connect_to_db(host_name, dbname, port, username, password):
    try:
        conn = ps.connect(host=host_name, dbname=dbname, user=username, password=password, port=port)
    except ps.OperationalError as e:
        raise e
    else:
        print('Connected!')
    return conn
        
```

```python
#create table
def create_table(curr):

    create_table_command = ("""CREATE TABLE IF NOT EXISTS data_val (
        sk_id_curr INTEGER,
        payment_rate FLOAT NOT NULL,
        ext_source_2 FLOAT NOT NULL,
        days_birth INTEGER NOT NULL,
        ext_source_3 FLOAT NOT NULL,
        amt_annuity FLOAT NOT NULL,
        annuity_income_perc FLOAT NOT NULL,
        days_id_publish INTEGER NOT NULL,
        amt_goods_price FLOAT NOT NULL,
        amt_credit FLOAT NOT NULL,
        days_registration FLOAT NOT NULL,
        instal_days_entry_payment_max FLOAT NOT NULL,
        prev_cnt_payment_mean FLOAT NOT NULL,
        days_last_phone_change FLOAT NOT NULL,
        days_employed_perc FLOAT NOT NULL,
        buro_days_credit_max FLOAT NOT NULL,
        instal_amt_payment_sum FLOAT NOT NULL,
        income_credit_perc FLOAT NOT NULL,
        instal_amt_payment_min FLOAT NOT NULL,
        active_days_credit_enddate_min FLOAT NOT NULL,
        closed_days_credit_max FLOAT NOT NULL,
        buro_days_credit_enddate_max FLOAT NOT NULL,
        approved_days_decision_max FLOAT NOT NULL,
        instal_dbd_sum FLOAT NOT NULL,
        instal_dbd_max FLOAT NOT NULL,
        instal_days_entry_payment_mean FLOAT NOT NULL,
        buro_amt_credit_sum_max FLOAT NOT NULL,
        instal_amt_payment_max FLOAT NOT NULL,
        instal_amt_instalment_max FLOAT NOT NULL,
        instal_days_entry_payment_sum FLOAT NOT NULL,
        prev_amt_annuity_mean FLOAT NOT NULL
        )""")
    
    curr.execute(create_table_command)
    
```

```python
def insert_into_table(curr, sk_id_curr, payment_rate, ext_source_2, days_birth, ext_source_3, amt_annuity,
    annuity_income_perc, days_id_publish, amt_goods_price, amt_credit, days_registration,
    instal_days_entry_payment_max, prev_cnt_payment_mean, days_last_phone_change, days_employed_perc,
    buro_days_credit_max, instal_amt_payment_sum, income_credit_perc, instal_amt_payment_min, active_days_credit_enddate_min,
    closed_days_credit_max, buro_days_credit_enddate_max, approved_days_decision_max, instal_dbd_sum, instal_dbd_max,
    instal_days_entry_payment_mean, buro_amt_credit_sum_max, instal_amt_payment_max, instal_amt_instalment_max,
    instal_days_entry_payment_sum, prev_amt_annuity_mean):
    
    insert_into_data_val = ("""INSERT INTO data_val
    VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);""")
    
    row_to_insert = (sk_id_curr, payment_rate, ext_source_2, days_birth, ext_source_3, amt_annuity,
    annuity_income_perc, days_id_publish, amt_goods_price, amt_credit, days_registration,
    instal_days_entry_payment_max, prev_cnt_payment_mean, days_last_phone_change, days_employed_perc,
    buro_days_credit_max, instal_amt_payment_sum, income_credit_perc, instal_amt_payment_min, active_days_credit_enddate_min,
    closed_days_credit_max, buro_days_credit_enddate_max, approved_days_decision_max, instal_dbd_sum, instal_dbd_max,
    instal_days_entry_payment_mean, buro_amt_credit_sum_max, instal_amt_payment_max, instal_amt_instalment_max,
    instal_days_entry_payment_sum, prev_amt_annuity_mean)
    
    curr.execute(insert_into_data_val, row_to_insert)
```

```python
def append_from_df_to_db(curr,df):
    for i, row in df.iterrows():
        insert_into_table(curr, row['SK_ID_CURR'], row['PAYMENT_RATE'], row['EXT_SOURCE_2'], row['DAYS_BIRTH'],
                          row['EXT_SOURCE_3'], row['AMT_ANNUITY'], row['ANNUITY_INCOME_PERC'], row['DAYS_ID_PUBLISH'],
                          row['AMT_GOODS_PRICE'], row['AMT_CREDIT'], row['DAYS_REGISTRATION'],
                          row['INSTAL_DAYS_ENTRY_PAYMENT_MAX'], row['PREV_CNT_PAYMENT_MEAN'],
                          row['DAYS_LAST_PHONE_CHANGE'], row['DAYS_EMPLOYED_PERC'], row['BURO_DAYS_CREDIT_MAX'],
                          row['INSTAL_AMT_PAYMENT_SUM'], row['INCOME_CREDIT_PERC'],
                          row['INSTAL_AMT_PAYMENT_MIN'], row['ACTIVE_DAYS_CREDIT_ENDDATE_MIN'],
                          row['CLOSED_DAYS_CREDIT_MAX'], row['BURO_DAYS_CREDIT_ENDDATE_MAX'],
                          row['APPROVED_DAYS_DECISION_MAX'], row['INSTAL_DBD_SUM'], row['INSTAL_DBD_MAX'],
                          row['INSTAL_DAYS_ENTRY_PAYMENT_MEAN'], row['BURO_AMT_CREDIT_SUM_MAX'],
                          row['INSTAL_AMT_PAYMENT_MAX'], row['INSTAL_AMT_INSTALMENT_MAX'],
                          row['INSTAL_DAYS_ENTRY_PAYMENT_SUM'], row['PREV_AMT_ANNUITY_MEAN'])
```
